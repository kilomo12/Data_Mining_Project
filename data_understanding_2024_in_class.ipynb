```python
for v in df['class'].unique():
    plt.hist(df[df['class'] == v]['petal length'], 
             bins=np.arange(0, 8, 0.25),
             label=v,
             alpha=0.6
            )
plt.title('petal length')
plt.legend()
plt.show()
```

```python
feature = 'sepal width'
for v in df['class'].unique():
    plt.hist(df[df['class'] == v][feature], 
             bins=np.arange(0, 8, 0.25),
             label=v,
             alpha=0.6
            )
plt.title(feature)
plt.legend()
plt.show()
```
